```python
# Methods: are like functions which define what the class does 
# Attributes: are like general characteristics of the class, something like what it has
# __init__: special method used to initialize attributes of an object 
# Class object attributes: Values that are same for all instances of a class. No need to pass it as an argument every time we create a instance of the object

class Dog: # This is how we define the name of the custom object
    species = "mammals" # class object attribute
    def __init__(self,breed): 
        self.breed = breed # Attribute 
    def bark(self): # Method
        print('Whaff!')
              
timmy = Dog(breed = "pug") # instance of the Dog class 
print(type(timmy))
print(timmy.breed)
print(timmy.bark())
print(timmy.species) #printing out class object instance values


class Circle:
    pi = 3.14 # class object instance 

    def __init__(self,radius):
        self.radius = radius

    def circumference(self):
        return 2 * self.pi * self.radius
    
    def area(self):
        return self.pi*(self.radius*self.radius)
    
    def setRadius(self,new_radius):
        self.radius = new_radius 
        return self.radius 
    
x = Circle(radius = 3)

print(x.radius)
print(x.circumference())
print(x.area())
print(x.setRadius(new_radius = 5))


class Line: 
    def __init__(self,coor1,coor2):
        self.coor1 = coor1
        self.coor2 = coor2

    def distance(self):
        x1,y1 = self.coor1
        x2,y2 = self.coor2 
        return ((x2-x2)**2+(y2-y1)**2)**0.5
    
    def slope(self):
        x1,y1 = self.coor1
        x2,y2 = self.coor2
        return (y2-y1)/(x2-x1)
    
coordinate1 = (3,2)
coordinate2 = (8,10)
li = Line(coordinate1,coordinate2)
li.distance()
li.slope()
```

```python
# Challenge Exercise 

class Account:
    def __init__(self,owner,balance):
        self.owner = owner 
        self.balance = balance

    def deposit(self,amt):
        self.balance += amt 
        print("Deposit Accepted")


    def withdraw(self,amt):
        if self.balance>= amt:
            self.balance -= amt
            print("Withdrwal Accepted")
        else:
            print("Funds Unavailable!")

    def __str__(self):
        return f"Account Owner: {self.owner}\nAccount Balance: ${self.balance}"

acc1 = Account('Nehal',100)

print(acc1)
acc1.owner
acc1.balance

acc1.deposit(50)

acc1.withdraw(75)

acc1.withdraw(500)
```

```python
# Inheritence 
# Build derived/child classes using parent or base class 
# This is used to reuse code and make it less complex

class Animal:
    def __init__(self):
        print("Animal created")
    def whoAmI(self):
        print("Animal")
    def eat(self):
        print("Eating")

class Dog(Animal):
    def __init__(self):
        Animal.__init__(self)
        print("Dog Created")

    def WhoAmI(self):
        print("Dog")

    def bark(self):
        print("Whoof!")

timmy = Dog()

timmy.eat()
timmy.bark()
timmy.WhoAmI()
```

```python
#Polymorphism 
# When 2 functions of diffent objects have a same method how we could possibly output the values of method together. 
# one possible way would be to define an abstract class(class which will never be instanciated)

class Dog:
    def __init__(self,breed):
        self.breed = breed
    def bark(self):
        print("Whoof!") 

class Cat:
    def __init__(self,breed):
        self.breed = breed 

    def bark(self):
        print("Moew!")  

timmy = Dog(breed = "pug")
laila = Cat(breed = "cute cat")

# We can print the value of methods as 
timmy.bark()
laila.bark()

for pet in [timmy,laila]:
    pet.bark()

def pet_bark(pet):
    print(pet.bark())
```

```python
# Using Abstract Class 

class Animal:
    def __init__(self,name):
        self.name = name

    def speak(self):
        raise NotImplementedError("Subclass must implement abstract method")

class Dog(Animal):
    def speak(self):
        return self.name + ' says whoof!'
    
class Cat(Animal):

    def speak(self):
        return self.name + ' says meow!'
    
timmy = Dog("timmy")
laila = Cat("laila")

print(timmy.speak())
print(laila.speak())
```

```python
# Special methods 

class Book:
    def __init__(self,author,book_name,num_of_pages):
        print("Book is created!")
        self.author = author
        self.book_name = book_name
        self.num_of_pages = num_of_pages

    def __str__(self):
        return(f"Title: {self.book_name}!,author:{self.author},pages:{self.num_of_pages}")

    def __len__(self):
        return self.num_of_pages
    
    def __del__(self):
        print("Book is destroyed!")


book = Book("Python Rocks!","Nehal Kumar",286)

print(book)
print(len(book))
del book
```
