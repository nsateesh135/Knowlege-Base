```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Functional Programming 
# Single plot on the canvas 
# Mention colour of the plot as third argument : 'r' for red
y = np.linspace(0,5,11)
x = y **2 
plt.plot(x,y,'r')
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
plt.show()
```

```python
# Plotting multiple plots on the same canvas 
# plt.subplot(num_rows,num_columns,plot_number)
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented Method
# fig.add_axes([left,bottom,width,height])
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.43,0.3])

axes1.plot(y,x)
axes2.plot(x,y)

axes2.set_title("Smaller Plot")
axes1.set_title("Larger Plot")
```

```python
# Plotting subplots using object oriented method 
# fig,axes : We are conducting tuple en-packing 
# axes: is iterable 
fig,axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x,y)
axes[0].set_title("First Plot")
axes[1].plot(y,x)
axes[1].set_title("Second Plot")

#plt.tight_layout(): takes care of any overlapping plots
# Always add this at the end of the plot statements 
plt.tight_layout()
```

```python
# figsize and dpi(dots per inch) with single plot 
fig = plt.figure(figsize=(5,2),dpi = 1000)
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
# Using figsize and dpi with subplots
fig,axes = plt.subplots(nrows = 2 , ncols = 1, figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
# How to save a visualization ?
fig.savefig("name_of_the_fig.png",dpi = 1000)
```

```python
# Working with Legends 
# loc = 0 : best 
# loc = 1 : Upper Right
# loc = 2 : Upper Left 
# loc = 3 : Lower Right
# loc = 10 : Center
# If none of the location codes satify we can put in our own location in the form (left[0.1],bottom[0.1])
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label = "x squared")
ax.plot(x,x**3,label = "x cubed")
# ax.legend(loc = 0)
ax.legend(loc=(0.1,0.1))
```

```python
# Personalizing chats 
# color = 'red' , to set color (we can also pass hex codes)
# linewidth = 2 , to set line width
# alpha = 0.5 , to set transparency
# linestyle = '--', to set linestyle
# marker = 'o' , to indicate data points on the chat
# markersize = 20, used to set the size of the marker
# markerfacecolor = 'yellow', used to set the color of the data points
# markeredgewidth = 3 , used to set width of the data points
# markeredgecolor = 'green', used to set the marker edge color
fig = plt.figure(figsize=(3,2))
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y,color='red',linewidth = 1,alpha = 0.5,linestyle= '--',marker ='o',markersize = 20,
        markerfacecolor = 'yellow',markeredgewidth = 3,markeredgecolor = 'green')

# Set x and y limit 
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
