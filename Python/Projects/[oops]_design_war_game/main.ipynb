```python
# Create a dictionary of card number and value 
import random
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
# Create a card class
# TODO1: Set attributes as suit and rank 
# TODO2: Set value attribute based on defined global values dictionary
# TODO3: Test values 
class Card:

    def __init__(self,suit,rank):
        self.suit = suit 
        self.rank = rank 
        self.value = values[self.rank]

    def __str__ (self):
        return self.rank + " of " + self.suit
    
# Test card class
# two_of_hearts = Card("Hearts","Two")
# print(two_of_hearts.suit)
# print(two_of_hearts.rank)
# print(two_of_hearts.value)
# print(two_of_hearts)
```

```python
# Build the deal class 
# The deal class should build all values for all suits i.e. a deck of 52 cards

class Deal:

    def __init__(self):
        self.all_cards = []

        for suit in suits:
            for rank in ranks:
                self.all_cards.append(Card(suit,rank))

    def shuffle(self):
        random.shuffle(self.all_cards)

    def deal_one(self):
        return self.all_cards.pop()
    
# Test Deck class
# my_deck = Deal()
#print(len(my_deck.all_cards)) # We have created a deck of 52 cards 
# deal_one_card = my_deck.deal_one()
```

```python
# Build a player class 
class Player:
    
    def __init__(self,name):
        self.name = name
        self.all_cards = []

    def remove_one(self):
        return self.all_cards.pop()
    
    def add_one(self,new_cards):
        if(type(new_cards) == type([])):
            return self.all_cards.extend(new_cards)
        else:
            return self.all_cards.append(new_cards)
    
    def __str__ (self):
        return f"Player {self.name} has {len(self.all_cards)} cards."
    
# Test player class
# player1 = Player(name = "Nehal")
# player1.add_one([deal_one_card,deal_one_card,deal_one_card])
```
